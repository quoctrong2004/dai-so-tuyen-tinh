#### Câu 1: Hãy viết chương trình bằng Python hoặc sử dụng thư viện numpy/sympy để: 


```python
#a) Tự chọn một ma trận khả nghịch 3x3. Sinh viên chứng minh ma trận đó khả nghịch (tồn tại ma trận nghịch đảo) 

import numpy as np
from sympy import Matrix

# a) Chọn một ma trận 3×3 khả nghịch
A = np.array([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
])

detA = round(np.linalg.det(A))
A_sym = Matrix(A.tolist())
detA_sym = A_sym.det()

print("det(A) =", detA)         
print("det(A) (Sympy) =", detA_sym)
if detA_sym == 0:
    raise ValueError("det(A) = 0, ma trận không khả nghịch!")

A_inv_np = np.linalg.inv(A)
A_inv_sym = A_sym.inv()

print("\nA^{-1} (NumPy) =")
print(np.round(A_inv_np, 6))    
print("\nA^{-1} (Sympy) =")
print(A_inv_sym)

I_np = A.dot(A_inv_np)
I_sym = (A_sym * A_inv_sym).evalf()

print("\nA·A^{-1} (NumPy) ≈")
print(np.round(I_np, 6))
print("\nA·A^{-1} (Sympy) =")
print(I_sym)
```

```python
# b) Nhập họ và tên hoặc mã số sinh viên (của sinh viên). 

text = input("Nhập họ và tên hoặc mã số sinh viên: ")
```

```python
#c) Mã hóa họ và tên hoặc mã số sinh viên (của sinh viên). 

text_bytes = [ord(char) for char in text]

while len(text_bytes) % 3 != 0:
    text_bytes.append(0) 

vectors = np.array(text_bytes).reshape(-1, 3).T  
encoded = A @ vectors

print("\nMã hóa (ma trận kết quả):")
print(encoded.astype(int))

```

```python
#d) Thực hiện giải mã với ma trận được chọn. 

decoded = np.linalg.inv(A) @ encoded

decoded_rounded = np.round(decoded).astype(int).T.flatten()
decoded_text = ''.join(chr(num) for num in decoded_rounded if num != 0)

print("\nGiải mã:")
print(decoded_text)
```

#### Câu 2: Tính toán phân số từ liên phân số (continued fractions) 


```python
# a. Xây dựng chương trình tính


def continued_fraction(c):
    n = len(c)
    p = [0] * n
    q = [0] * n

    p[0] = c[0]
    q[0] = 1

    if n > 1:
        p[1] = c[1] * c[0] + 1
        q[1] = c[1]

    for i in range(2, n):
        p[i] = c[i] * p[i - 1] + p[i - 2]
        q[i] = c[i] * q[i - 1] + q[i - 2]

    return p[-1], q[-1] 

c = [3, 1, 2] 
pn, qn = continued_fraction(c)
print(f"Phân số: {pn}/{qn}")
```
